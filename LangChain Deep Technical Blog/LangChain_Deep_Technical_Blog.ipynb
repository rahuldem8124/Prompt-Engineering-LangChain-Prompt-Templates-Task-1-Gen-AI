```python
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage

# We updated the model name to Groq's active 'Llama 3.3 70B' model!
chat_model = ChatGroq(model="llama-3.3-70b-versatile", temperature=0.7)

print("--- Basic LLM Call ---")
messages = [
    SystemMessage(content="You are a helpful Python expert."),
    HumanMessage(content="What is a list comprehension?")
]

response = chat_model.invoke(messages)
print(response.content)
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

print("\n--- Prompts & Chains (LCEL) ---")

# Define the Prompt Template
template = PromptTemplate(
    input_variables=["language", "task"],
    template="Write a short, 2-line {language} script to {task}."
)

# Create a modern LCEL Chain: Prompt -> Model -> Output Parser
chain = template | chat_model | StrOutputParser()

# Invoke the chain
result = chain.invoke({"language": "Python", "task": "print the current date"})
print(result)
```

```python
# Install the missing community package for LangChain's memory tools
!pip install -q langchain-community

from langchain_community.memory.buffer import ConversationBufferMemory
from langchain.chains import LLMChain
from langchain_core.prompts import PromptTemplate

print("\n--- Conversation Memory ---")

# Setup memory to remember past messages
memory = ConversationBufferMemory(memory_key="chat_history")

prompt = PromptTemplate(
    input_variables=["chat_history", "question"],
    template="History:\n{chat_history}\nUser: {question}\nAI:"
)

# Using LLMChain for memory demonstration
conversation = LLMChain(
    llm=chat_model,
    prompt=prompt,
    memory=memory
)

# Turn 1
print("User: Hi, I am Rahul.")
print("AI:", conversation.invoke({"question": "Hi, I am Rahul."})['text'])

# Turn 2 (Testing if the model remembers your name)
print("User: What is my name?")
print("AI:", conversation.invoke({"question": "What is my name?"})['text'])
```

```python
from langchain.agents import load_tools, initialize_agent, AgentType

print("\n--- Agents and Tools ---")

# Load the Wikipedia tool and bind it to our Groq model
tools = load_tools(["wikipedia"], llm=chat_model)

# Initialize the Agent
agent = initialize_agent(
    tools,
    chat_model,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
    handle_parsing_errors=True # Helps the agent recover if it formats the tool request slightly wrong
)

# Test the agent with a query that requires external Wikipedia knowledge
print("Testing Agent...")
agent.invoke("What is the capital of France according to Wikipedia?")
```
