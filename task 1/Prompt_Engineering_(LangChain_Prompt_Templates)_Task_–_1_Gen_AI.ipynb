```python
print("--- TASK 1: Replace Hardcoded Prompts ---")

# The old way (Hardcoded / f-strings):
# def explain_topic(topic): return f"Explain {topic} in simple terms for beginners"

# The LangChain way:
task1_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Test the template
topic_input = "Generative AI"
formatted_prompt_1 = task1_template.format(topic=topic_input)

print("Original Input Topic:", topic_input)
print("Generated Prompt:", formatted_prompt_1)
print()
```

```python
print("--- TASK 2: Multi-Input Prompt System ---")

# Define the template with multiple variables
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone."
)

# Define the test cases from the assignment
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

# Run the test cases
for i, test in enumerate(test_cases, 1):
    result = multi_input_template.format(**test)
    print(f"Test Case {i}: {result}")
print()
```

```python
print("--- TASK 3: Prompt Variations Engine ---")

# Create 3 different PromptTemplates
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step."
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}."
)

storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story."
)

# Run for topic = Machine Learning
target_topic = "Machine Learning"

print("Teaching Variation:", teaching_template.format(topic=target_topic))
print("Interview Variation:", interview_template.format(topic=target_topic))
print("Storytelling Variation:", storytelling_template.format(topic=target_topic))
print()
```

```python
print("--- TASK 4: ChatPromptTemplate System ---")

# Build a chat-based prompt system
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}. Behave and answer exactly as this role would."),
    ("user", "Explain the following topic to me: {topic}")
])

# Test with the requested example
role_input = "teacher"
topic_input = "Neural Networks"

chat_prompt_output = chat_template.format_messages(role=role_input, topic=topic_input)

# Display the roles and their corresponding generated content
for message in chat_prompt_output:
    # message.type will show 'system' or 'human' (user)
    print(f"Role: [{message.type.upper()}] -> Content: {message.content}")
print()
```

```python
print("--- TASK 5: Input Validation Layer ---")

def validate_inputs(audience, tone):
    allowed_audiences = ["beginner", "intermediate", "expert"]
    allowed_tones = ["formal", "casual", "fun"]

    # Assign default values if validation fails
    final_audience = audience if audience in allowed_audiences else "beginner"
    final_tone = tone if tone in allowed_tones else "casual"

    # Optional: Print a warning if defaults were used
    if final_audience != audience or final_tone != tone:
        print(f"⚠️ Validation Warning: Invalid inputs detected. Defaulting to Audience: '{final_audience}', Tone: '{final_tone}'.")

    return final_audience, final_tone

# Test Valid Inputs
print("Test 1 (Valid):", validate_inputs("expert", "formal"))

# Test Invalid Inputs (Should trigger defaults)
print("Test 2 (Invalid):", validate_inputs("toddler", "angry"))
print()
```

```python
print("--- TASK 6: Prompt Generator App ---")

def generate_prompt(topic, audience, tone, style):
    # Step 1: Validate inputs
    valid_audience, valid_tone = validate_inputs(audience, tone)

    # Step 2: Define the template (matching the expected output format)
    app_template = PromptTemplate(
        input_variables=["topic", "audience", "tone", "style"],
        template="Explain {topic} for {audience} in a {tone} {style} style."
    )

    # Step 3: Generate and return the prompt
    return app_template.format(
        topic=topic,
        audience=valid_audience,
        tone=valid_tone,
        style=style
    )

# Run the example from the assignment
generated_output = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)

print("Generated App Prompt:", generated_output)
print()
```

```python
print("--- TASK 7: Template Reusability Test ---")

# ONE template to rule them all
reusable_template = PromptTemplate(
    input_variables=["action", "subject", "context"],
    template="{action} {subject} {context}."
)

# 5 completely different inputs
reusable_inputs = [
    {"action": "Summarize", "subject": "the article", "context": "in 3 bullet points"},
    {"action": "Translate", "subject": "the greeting", "context": "into French"},
    {"action": "Write a Python script for", "subject": "data cleaning", "context": "using pandas"},
    {"action": "Generate 5 marketing taglines for", "subject": "a new sports drink", "context": "targeting athletes"},
    {"action": "Debug", "subject": "this SQL query", "context": "and explain the error"}
]

# Run the single template 5 times
for i, inputs in enumerate(reusable_inputs, 1):
    print(f"Reuse {i}: {reusable_template.format(**inputs)}")
```
